```python
import pandas as pd
```

```python
# Loading each CSV file
file_paths = {
    "departments": "/users/jt/Data Analytics/Module 9 Challenge/data/departments.csv",
    "dept_emp": "/users/jt/Data Analytics/Module 9 Challenge/data/dept_emp.csv",
    "dept_manager": "/users/jt/Data Analytics/Module 9 Challenge/data/dept_manager.csv",
    "employees": "/users/jt/Data Analytics/Module 9 Challenge/data/employees.csv",
    "salaries": "/users/jt/Data Analytics/Module 9 Challenge/data/salaries.csv",
    "titles": "/users/jt/Data Analytics/Module 9 Challenge/data/titles.csv"
}
```

```python
# Storing dataframes for inspection
dataframes = {}

# Loading the data and inspect the first few rows
for name, path in file_paths.items():
    dataframes[name] = pd.read_csv(path)
    print(f"--- {name.upper()} ---")
    print(dataframes[name].head(), "\n")
    print(dataframes[name].info(), "\n")
```

```python
# Returning a summary of each file for understanding relationships
data_summary = {name: {"columns": df.columns.tolist(), "sample": df.head()} for name, df in dataframes.items()}
data_summary
```

## COVID-19 Statistics, Draft 001
- This notebook will contain data regarding COVID deaths and hospitalizations
- Data is sourced from the CDC.
- https://covid.cdc.gov/covid-data-tracker/

```python
# Import dependencies
import pandas as pd
from pathlib import Path
```

```python
# Read CSV file, store into DataFrame
covid_death_csv = ("../Data/Provisional_COVID-19_Death_Counts_by_Week_Ending_Date_and_State_CDC.csv")
covid_df = pd.read_csv(covid_death_csv)
```

```python
# Clean DataFrame, remove unnecessary columns, filter Group by month, reset index
covid_df = covid_df[['Start Date','End Date','Group','Year','Month','State','COVID-19 Deaths','Total Deaths']]
covid_df = covid_df.loc[covid_df['Group']=='By Month']
covid_df = covid_df.reset_index(drop=True)
```

```python
covid_df.info()
```

```python
# Fill NaN values in 'COVID-19 Deaths' column with 0
covid_df = covid_df.fillna(0)
```
